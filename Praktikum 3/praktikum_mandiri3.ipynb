```python
# Mengimport Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
#import data dan membaca file csv
df = pd.read_csv('StudentPerformanceFactors.csv')
df
```

```python
df.info()
```

```python
# Mengecek apakah ada missing values atau nggak
df.isnull().sum()
```

```python
# Dropping unecessary columns atau
# Hapus data yang tidak diperlukan karena seluruh data memiliki nilai kosong
columns_to_drop =['Teacher_Quality','Parental_Education_Level','Distance_from_Home']
df.drop(columns=columns_to_drop, inplace=True) # inplace=False atau tidak disebutkan, perintah ini akan mengembalikan DataFrame baru, sementara df asli tidak berubah.
df
```

1. Statistik Deskriptif
Kita akan mulai dengan menghitung statistik deskriptif seperti mean, median, modus, max, min, quantiles, outliers, dan variance untuk kolom Hours_Studied.

```python
# Tampilkan statistik deskriptif
print(df.describe())
```

```python
# Hitung statistik deskriptif hours_studied
mean_hours = df['Hours_Studied'].mean()
median_hours = df['Hours_Studied'].median()
mode_hours = df['Hours_Studied'].mode()[0]  # Ambil modus pertama
max_hours = df['Hours_Studied'].max()
min_hours = df['Hours_Studied'].min()
quantiles_hours = df['Hours_Studied'].quantile([0.25, 0.5, 0.75])  # 25th, 50th, 75th percentiles
variance_hours = df['Hours_Studied'].var()

# Menampilkan hasil
print(f'Mean: {mean_hours}')
print(f'Median: {median_hours}')
print(f'Modus: {mode_hours}')
print(f'Max: {max_hours}')
print(f'Min: {min_hours}')
print(f'Quantiles:\n{quantiles_hours}')
print(f'Variance: {variance_hours}')
```

2. Deteksi Outliers
Kita bisa mendeteksi outliers dengan menggunakan metode IQR (Interquartile Range).

```python
# Menghitung IQR
Q1 = df['Hours_Studied'].quantile(0.25)
Q3 = df['Hours_Studied'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menemukan outliers
outliers = df[(df['Hours_Studied'] < lower_bound) | (df['Hours_Studied'] > upper_bound)]
print(f'Jumlah outliers: {outliers.shape[0]}')
print(outliers[['Hours_Studied']])  # Menampilkan outliers
```

3. Visualisasi Sebaran Data
Untuk lebih memahami sebaran data, kita bisa menggunakan histogram dan boxplot.

```python
# Histogram untuk Hours_Studied
plt.figure(figsize=(10, 6))
sns.histplot(df['Hours_Studied'], bins=10, kde=True)
plt.title('Distribusi Jam Belajar')
plt.xlabel('Jam Belajar')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Boxplot untuk Hours_Studied
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Hours_Studied'])
plt.title('Boxplot Jam Belajar')
plt.xlabel('Jam Belajar')
plt.show()
```

```python
#Memfilter Outlier
df = df[(df['Hours_Studied'] >= lower_bound) & (df['Hours_Studied'] <= upper_bound)]
```

```python
#Mengecek ada missing values atau nggak
df.isna().sum()
```

```python
#mengecek data duplicate
duplicates = df[df.duplicated()]
print(duplicates)
```

```python

# Boxplot untuk Hours_Studied
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Hours_Studied'])
plt.title('Boxplot Jam Belajar')
plt.xlabel('Jam Belajar')
plt.show()
```

```python
#Mengecek boxplot dari 3 kolom
df.boxplot(column=[ 'Hours_Studied','Tutoring_Sessions','Exam_Score' ])
plt.title('Boxplot of Tutoring_Sessions,Exam_Score with Outliers')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Menghitung Q1 (kuartil pertama) dan Q3 (kuartil ketiga)
Q1 = df[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']].quantile(0.25)
Q3 = df[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']].quantile(0.75)
IQR = Q3 - Q1

# Mendefinisikan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menghapus outlier
df_cleaned = df[~((df[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']] < lower_bound) | (df[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']] > upper_bound)).any(axis=1)]
```

```python
# Menggunakan MinMaxScaler untuk mengubah nilai ke dalam rentang antara 0 dan 1
scaler = MinMaxScaler()
df_cleaned.loc[:, ['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']] = scaler.fit_transform(df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']])
```

```python
# Membuat ulang boxplot setelah menangani outliers
df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']].plot(kind='box', title='Boxplot after handling Outliers')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Boxplot visualisasi setelah normalisasi dan penanganan outliers
plt.figure(figsize=(6, 3))

sns.boxplot(data=df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']])

plt.title('Boxplot after Handling Outliers and Normalization')
plt.ylabel('Values')
plt.xlabel('Variables')
plt.show()
```

```python
# Menggunakan StandardScaler untuk menstandarkan data dengan mean = 0 dan standard deviation = 1
scaler = StandardScaler()
df_cleaned.loc[:, ['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']] = scaler.fit_transform(df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']])
```

```python
# Membuat ulang boxplot setelah menangani outliers
df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']].plot(kind='box', title='Boxplot after handling Outliers')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Boxplot visualisasi setelah normalisasi dan penanganan outliers
plt.figure(figsize=(6, 3))

sns.boxplot(data=df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']])

plt.title('Boxplot after Handling Outliers and Normalization')
plt.ylabel('Values')
plt.xlabel('Variables')
plt.show()
```

```python
# Menggunakan MaxAbsScaler untuk mengubah data ke dalam rentang antara -1 dan 1 (berdasarkan nilai absolut terbesar)
scaler = MaxAbsScaler()
df_cleaned.loc[:, ['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']] = scaler.fit_transform(df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']])
```

```python
# Membuat ulang boxplot setelah menangani outliers
df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']].plot(kind='box', title='Boxplot after handling Outliers')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Boxplot visualisasi setelah normalisasi dan penanganan outliers
plt.figure(figsize=(6, 3))

sns.boxplot(data=df_cleaned[['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']])

plt.title('Boxplot after Handling Outliers and Normalization')
plt.ylabel('Values')
plt.xlabel('Variables')
plt.show()
```

Visualisasi data lebih lanjut

```python
# Menghitung rata-rata skor ujian berdasarkan jam belajar
average_scores = df.groupby('Hours_Studied')['Exam_Score'].mean().reset_index()

# Membuat diagram garis
plt.figure(figsize=(10, 5))
plt.plot(average_scores['Hours_Studied'], average_scores['Exam_Score'], 
         marker='o', color='blue', linestyle='-', linewidth=2)

# Menambahkan detail dari statistik deskriptif ke dalam judul
plt.title('Hubungan antara Jam Belajar dan Rata-rata Skor Ujian', fontsize=14)
plt.xlabel('Jam Belajar', fontsize=12)
plt.ylabel('Rata-rata Skor Ujian', fontsize=12)

# Mengatur rentang sumbu y jika perlu, misalnya dari 0 hingga nilai maksimum yang sesuai
plt.ylim(0, 100)  # Sesuaikan dengan rentang skor ujian yang diharapkan

# Menampilkan semua jam belajar di sumbu x
plt.xticks(average_scores['Hours_Studied'])

plt.grid(True)
plt.show()
```

```python
# Menghitung rata-rata skor ujian berdasarkan jam belajar
average_scores = df.groupby('Hours_Studied')['Exam_Score'].mean().reset_index()

# Membuat diagram batang horizontal
plt.figure(figsize=(10, 10))
plt.barh(average_scores['Hours_Studied'], average_scores['Exam_Score'], 
         color='skyblue')

# Menambahkan detail dari statistik deskriptif ke dalam judul
plt.title('Jam Belajar & Rata-rata Skor Ujian', fontsize=14)
plt.xlabel('Rata-rata Skor Ujian', fontsize=12)
plt.ylabel('Jam Belajar', fontsize=12)

# Mengatur rentang sumbu x (karena diagram batang horizontal) jika perlu
plt.xlim(0, 100)  # Sesuaikan dengan rentang skor ujian yang diharapkan

# Menampilkan semua jam belajar di sumbu y
plt.yticks(average_scores['Hours_Studied'])

plt.grid(True, axis='x')
plt.show()
```
