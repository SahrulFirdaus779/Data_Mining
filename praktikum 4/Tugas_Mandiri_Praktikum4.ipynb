## Top 10 frequently sold products

```python
# Mendefinisikan fungsi untuk membuat plot distribusi
def ditribution_plot(x, y, name=None, xaxis=None, yaxis=None):
    # Membuat objek Figure dari plotly.graph_objects dengan data bar
    fig = go.Figure([
        go.Bar(x=x, y=y)  # Menambahkan grafik batang dengan sumbu x dan y yang ditentukan
    ])

    # Memperbarui tata letak grafik
    fig.update_layout(
        title_text=name,  # Menentukan judul grafik menggunakan parameter 'name'
        xaxis_title=xaxis,  # Menentukan judul sumbu x menggunakan parameter 'xaxis'
        yaxis_title=yaxis  # Menentukan judul sumbu y menggunakan parameter 'yaxis'
    )

    # Menampilkan grafik
    fig.show()  # Menampilkan plot yang telah dibuat
```

```python
# Menghitung jumlah kemunculan setiap produk dalam kolom 'itemDescription'
x = df['itemDescription'].value_counts()  # Menggunakan value_counts() untuk mendapatkan frekuensi setiap produk

# Mengurutkan hasil berdasarkan jumlah kemunculan secara menurun
x = x.sort_values(ascending=False)  # Mengurutkan hasil dengan nilai tertinggi di atas

# Mengambil 10 produk teratas berdasarkan frekuensi kemunculan
x = x[:10]  # Memilih 10 produk teratas dari hasil yang sudah diurutkan

# Memanggil fungsi ditribution_plot untuk membuat grafik batang
ditribution_plot(x=x.index, y=x.values, yaxis="Count", xaxis="Products")  # Menggunakan index sebagai sumbu x dan frekuensi sebagai sumbu y
```

## Transaksi
Note: Jika seorang pelanggan membeli beberapa produk pada hari yang sama, pembelian tersebut akan dihitung sebagai satu transaksi saja, bukan beberapa transaksi terpisah. Dengan kata lain, semua produk yang dibeli oleh pelanggan pada hari yang sama dianggap bagian dari satu pembelian atau satu transaksi.

```python
# Misalkan df dan all_products sudah didefinisikan
# Pastikan all_products hanya berisi kolom yang ada di df
valid_products = [product for product in all_products if product in df.columns]

# Kelompokkan dan jumlahkan
records = df.groupby(["Member_number", "Date"])[valid_products].sum()

# Reset index
records = records.reset_index()

# Memastikan semua kolom yang ingin diambil ada dalam records
final_records = records[["Member_number", "Date"] + valid_products]  # Menambahkan kolom Member_number dan Date kembali
```

```python
# Mendefinisikan fungsi untuk mengganti nilai non-nol dengan nama produk
def get_Pnames(x):
    for product in all_products:  # Mengiterasi melalui semua produk yang ada dalam list all_products
        if x[product] > 0:  # Memeriksa apakah nilai untuk produk tersebut lebih dari nol
            x[product] = product  # Jika ya, ganti nilai dengan nama produk
    return x  # Mengembalikan baris yang telah dimodifikasi

# Menerapkan fungsi get_Pnames pada setiap baris dalam DataFrame records
records = records.apply(get_Pnames, axis=1)  # Menggunakan apply untuk menerapkan fungsi pada setiap baris (axis=1)

# Menampilkan 5 baris pertama dari DataFrame records yang telah dimodifikasi
records.head()  # Menampilkan 5 baris pertama untuk memeriksa hasil
```

```python
# Menghitung dan menampilkan total transaksi yang ada dalam DataFrame records
print("total transactions: {}".format(len(records)))  # Menggunakan format untuk menyisipkan jumlah transaksi ke dalam string
```

```python
# Menghapus nilai nol untuk membangun transaksi, hanya mengambil item mulai dari index 2
x = records.values  # Mengambil nilai dari DataFrame 'records'
# Membuat list transaksi dengan menghilangkan item nol dan hanya mengambil item dari index 2
x = [sub[2:][sub[2:] != 0].tolist() for sub in x if sub[2:][sub[2:] != 0].tolist()]
transactions = x  # Menyimpan hasilnya dalam variabel 'transactions'

# Memastikan semua elemen dalam transaksi adalah string
transactions = [list(map(str, transaction)) for transaction in transactions]  # Mengonversi setiap item dalam transaksi menjadi string

# Menampilkan transaksi untuk verifikasi
print(transactions[:10])  # Menampilkan 10 transaksi pertama untuk diperiksa
```

## Association Rules

```python
# Mencari aturan asosiasi menggunakan algoritma apriori
rules = apriori(transactions,
                min_support=0.00030,  # Minimum dukungan 0.00030
                min_confidance=0.05,   # Minimum kepercayaan 5%
                min_lift=3,            # Minimum lift 3
                min_length=2,          # Minimum panjang aturan 2 item
                target="rules")        # Target keluaran adalah aturan

# Mengonversi hasil aturan asosiasi menjadi list
association_results = list(rules)  # Menyimpan hasil dalam bentuk list
```

```python
# Iterasi melalui setiap item dalam hasil aturan asosiasi
for item in association_results:

    pair = item[0]  # Mengambil pasangan item dari hasil aturan
    items = [x for x in pair]  # Mengonversi pasangan item menjadi list

    # Menampilkan aturan asosiasi dalam format "Item1 -> Item2"
    print("Rule: " + items[0] + " -> " + items[1])

    # Menampilkan nilai dukungan (support) dari aturan
    print("Support: " + str(item[1]))

    # Menampilkan nilai kepercayaan (confidence) dari aturan
    print("Confidence: " + str(item[2][0][2]))

    # Menampilkan nilai lift dari aturan
    print("Lift: " + str(item[2][0][3]))

    # Menampilkan garis pemisah untuk visualisasi
    print("=====================================")
```
